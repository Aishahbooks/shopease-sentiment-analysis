# 08 — Real-World Validation
## ShopEase Europe | Sentiment Analysis Project 
**Objective:** Validate our trained models against genuine customer 
reviews from a real-world multilingual dataset, addressing the 
template-based limitation identified in the ShopEase dataset. This 
notebook tests whether our models generalise beyond the synthetic 
training data.

```python
import os
import pandas as pd
from datasets import load_dataset
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded successfully")
```

## Download Real-World Sample
Pulling a small sample of genuine customer reviews across our four 
core languages from the Amazon Multilingual Reviews dataset on 
Hugging Face.

```python
# Load a small streaming sample to avoid downloading the full dataset
languages = ['en', 'fr', 'de', 'es']
samples = []

for lang in languages:
    dataset = load_dataset('mteb/amazon_reviews_multi', lang, split='test', streaming=True)
    lang_samples = list(dataset.take(15))
    for item in lang_samples:
        samples.append({
            'language': lang,
            'review_text': item['text'],
            'star_rating': item['label'] + 1  # dataset labels are 0-4, convert to 1-5
        })

real_world_df = pd.DataFrame(samples)
print(f"Downloaded {len(real_world_df)} real reviews across {len(languages)} languages")
print(f"\nSample by language:")
print(real_world_df['language'].value_counts())
print(f"\nFirst few reviews:")
print(real_world_df.head())
```

## Real-World Sample Confirmed

60 genuine customer reviews downloaded from the Amazon Multilingual 
Reviews dataset, 15 each across English, French, German and Spanish. 
Star ratings range from 1 to 5, providing ground truth for sentiment 
validation.

Initial inspection confirms this sample exhibits the linguistic 
variability and authentic complaint patterns absent from the 
ShopEase synthetic dataset. Exclamation marks, capitalisation 
for emphasis, varied sentence structure and genuine product-specific 
detail rather than templated phrasing.
